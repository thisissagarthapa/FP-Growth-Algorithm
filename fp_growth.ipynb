## FP-Growth Algorithm

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("amazon_transactions.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=["Transaction ID","Purchase Date"],inplace=True)
```

```python
dataset.head(3)
```

```python

market = []
for i in range(dataset.shape[0]):  # Loop through each row
    cus = []
    for j in dataset.columns:  # Loop through each column
        cus.append(dataset[j][i])  # Access the value in column 'j' and row 'i'
    market.append(cus)  # Add the row data to 'market'

```

```python
list=[]
for i in market:
    for j in i:
        list.append(j)
```

```python
import collections
```

```python
col=collections.Counter(list)
```

```python
d={"item_name":col.keys(),"values":col.values()}
```

```python
df1=pd.DataFrame(d).sort_values(by=["values"],ascending=False)
df1
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(df1)
```

```python
df2=pd.DataFrame(tr.transform(df1),columns=tr.columns_)
df2
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fp=fpgrowth(df2,use_colnames=True,max_len=3,min_support=0.02)
```
